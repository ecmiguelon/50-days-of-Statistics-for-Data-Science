**Key Terms:**

- Treatment: Something (drug, price, web headline) to which a subject is exposed.

- Treatment group: A group of subjects exposed to a specific treatment.

- Control group: A group of subjects exposed to no (or standard) treatment.

- Randomization: The process of randomly assigning subjects to treatments.

- Subjects: The items (web visitors, patients, etc.) that are exposed to treatments.

- Test statistic: The metric used to measure the effect of the treatment.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
```

```python
#load the dataset
data = pd.read_csv("D:\statistics\Dataset\data_abtesting.csv")
print(data.shape)
data.head()
```

```python
#plotting the distplot for Conversion_A column from the dataset.
sns.distplot(data.Conversion_A)
```

```python
#plotting the distplot for Conversion_B column from the dataset.
sns.distplot(data.Conversion_B)
```

```python
t_stat, p_val= stats.ttest_ind(data.Conversion_B,data.Conversion_A)
t_stat , p_val
```

On the calculation of the t-score, we get the t-score as .3787. and the p-value is 0.00036.

- Here, our p-value is less than the significance level i.e 0.05. 
- Hence, we can reject the null hypothesis. 
- This means that in our A/B testing, newsletter B is performing better than newsletter A. 
- So our recommendation would be to replace our current newsletter with B to bring more traffic on our website.
