**Chi-Square Distribution:**


- An important idea in statistics is departure from expectation, especially with respect to category counts. 

- Expectation is defined loosely as “nothing unusual or of note in the data” (e.g., no correlation between variables or predictable patterns). 

- This is also termed the “null hypothesis” or “null model”.

- The chi-square distribution is typically concerned with counts of subjects or items falling into categories.

- The chi-square statistic measures the extent of departure from what you would expect in a null model.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
  
# Using chisquare() method we are able to get the chi-square distribution and return the scalar numpy array by using this method.

data = np.random.chisquare(3, 1000)
  
count, bins, ignored = plt.hist(data, 10, density = True)
plt.show()
```

```python
#df - (degree of freedom)
#size - The shape of the returned array

x = np.random.chisquare(df=2, size=(2, 3))

print(x)
```

```python
sns.distplot(np.random.chisquare(df=1, size=1000), hist=False)

plt.show()
```
