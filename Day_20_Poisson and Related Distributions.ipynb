**Key Terms:**

- Lambda: The rate (per unit of time or space) at which events occur.

- Poisson distribution: The frequency distribution of the number of events in sampled units of time or space.

- Exponential distribution: The frequency distribution of the time or distance from one event to the next event.

- Weibull distribution: A generalized version of the exponential distribution in which the event rate is
allowed to shift over time.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#Poisson Distributions - Poisson Processes. 
#Poisson Processes are a type of random process that occur at a fixed rate.
sample = stats.poisson.rvs(5, size=100)

pd.Series(sample).plot.hist()
plt.show()
```

```python
#Exponential Distribution is also a special case of Poisson Distribution
sample = stats.expon.rvs(scale=20, size=100)

pd.Series(sample).plot.hist()
plt.show()
```

```python
#Weibull Distribution (also known as Cauchy Distribution).  
# It is a continuous distribution that is bounded above and below.
sample = stats.weibull_min.rvs(1.5, scale=5000, size=100)

pd.Series(sample).plot.hist()
plt.show()
```
