**Key Terms:**
1. Error:  The difference between a data point and a predicted or average value.

2. Standardize: Subtract the mean and divide by the standard deviation.

3. z-score: The result of standardizing an individual data point.

4. Standard normal: A normal distribution with mean = 0 and standard deviation = 1.

5. QQ-Plot: A plot to visualize how close a sample distribution is to a specified distribution,
e.g., the normal distribution.

QQ-Plot stands for Quantile-Quantile Plot.


```python
#importing libraries

import matplotlib.pyplot as plt
import scipy.stats as stats
```

1. What does Standardizing the data mean?

- It essentially means to subtract the mean from a data-point selected from the dataset and dividing the result by the standard deviation.

2. What is Standard Normal?

 - A normal distribution with mean =0 and std = 1

```python
#The package scipy has the function (scipy.stats.probplot) to create QQ-plots. 
# The argument dist specifies the distribution, which is set by default to the normal distribution.
fig, ax = plt.subplots(figsize=(8, 6))

norm_sample = stats.norm.rvs(size=1000)
stats.probplot(norm_sample, plot=ax)

plt.tight_layout()
plt.show()
```
